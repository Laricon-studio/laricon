```python
name = input("What is your name? ")

number_text = input("Enter a number: ")

if number_text.isdigit():
    number = int(number_text)
    result = number * 2
    print("Hello", name)
    print("Your number multiplied by 2 is:", result)
else:
    print("Please enter a valid number next time.")
```
